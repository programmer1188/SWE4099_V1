```python
#Split Independent and Dependent variables X y
X = data1.drop(columns=['Result'])
y = data1['Result']
```

```python
X.head()
```

```python
y
```

```python
#Feature Importance using Extra Tree Classifier
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X,y)

plt.figure(figsize=(10,10), facecolor='white')
col_imp = pd.Series(model.feature_importances_, index=X.columns)
col_imp.nlargest(30).plot(kind='barh')
plt.show()
```

```python
#Split the data into Test and Train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
X_train.shape, X_test.shape
```

```python
# Creating holders to store the model performance results
ML_Model = []
acc_test = []

#function to call for storing the results
def storeResults(model, a):
    if model not in ML_Model:
        ML_Model.append(model)
        acc_test.append(round(a, 3))
```

```python
#import all metrics to evaluate the model
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()

#Applying Cross Validation 
print(cross_val_score(log_reg, X, y, scoring='accuracy', cv = 10))
Accuracy = cross_val_score(log_reg, X, y, scoring='accuracy', cv = 10).mean()*100
print("\n AVG. Cross Validation Accuracy for LogisticRegression is: ",Accuracy)
print()

#logistic regression classifier fit
log_reg = log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)

#Accuracy
acc_log_reg = accuracy_score(y_test,y_pred)
print("Accuracy of Logistic Regression: ",acc_log_reg)
print()

#Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
print()

print(classification_report(y_test, y_pred))
```

```python
#Store Result in the list
storeResults('Logistic Regression', acc_log_reg)
```

```python
#Roc 
logit_roc_auc = roc_auc_score(y_test, log_reg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, log_reg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.3f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

```python
# "Support Vector Classifier"
from sklearn.svm import SVC  
svc = SVC(probability=True) 

#Applying Cross Validation 
print(cross_val_score(svc, X, y, scoring='accuracy', cv = 10))
Accuracy = cross_val_score(svc, X, y, scoring='accuracy', cv = 10).mean()*100
print("\n AVG. Cross Validation Accuracy for SVC is: ",Accuracy)
print()

#Support Vector classifier fit
svc = svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

#Accuracy
acc_svc = accuracy_score(y_test,y_pred)
print("Accuracy of SVC: ",acc_svc)
print()

#Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
print()

print(classification_report(y_test, y_pred))
```

```python
#Store Result in the list
storeResults('Support Vector Classifier', acc_svc)
```

```python
#Roc 
logit_roc_auc = roc_auc_score(y_test, svc.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, svc.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='SVC (area = %0.3f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('SVC_ROC')
plt.show()
```

```python
#Knn
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()

#Applying Cross Validation 
print(cross_val_score(knn, X, y, scoring='accuracy', cv = 10))
Accuracy = cross_val_score(knn, X, y, scoring='accuracy', cv = 10).mean()*100
print("\n AVG. Cross Validation Accuracy for Knn is: ",Accuracy)
print()

#knn model fit
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

#accuracy
acc_knn = accuracy_score(y_test, y_pred)
print("Accuracy of Knn: ",acc_knn)
print()

# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

#Classification Report
print(classification_report(y_test, y_pred))


```

```python
#Store Result in List
storeResults('knn', acc_knn)
```

```python
#Roc 
knn_roc_auc = roc_auc_score(y_test, knn.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, knn.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='knn (area = %0.3f)' % knn_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('knn_ROC')
plt.show()
```

```python
#Checking knn with various neighbours  
neighbors = np.arange(1, 9)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))
  
# Loop over K values
for i, k in enumerate(neighbors):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
      
    # Compute traning and test data accuracy
    train_accuracy[i] = knn.score(X_train, y_train)
    test_accuracy[i] = knn.score(X_test, y_test)
  
# Generate plot
plt.plot(neighbors, test_accuracy, label = 'Testing dataset Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training dataset Accuracy')
  
plt.legend()
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
#Fitting the model with neighbor = 1 
knn = KNeighborsClassifier(n_neighbors=1)
  
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

#accuracy
acc_knn = accuracy_score(y_test, y_pred)
print("Accuracy of Knn: ",acc_knn)
print()

# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

#Classification Report
print(classification_report(y_test, y_pred))
```

```python
#Store Result in List
storeResults('knn with n=1', acc_knn)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier

#Create a Random Forest Classifier
rf = RandomForestClassifier()

#Applying Cross Validation 
print(cross_val_score(rf, X, y, scoring='accuracy', cv = 10))
Accuracy = cross_val_score(rf, X, y, scoring='accuracy', cv = 10).mean()*100
print("\n AVG. Cross Validation Accuracy for RF is: ",Accuracy)
print()

#Train the model using the training sets
rf = rf.fit(X_train,y_train)

#Predict the response
y_pred = rf.predict(X_test)

# Model Accuracy
acc_rf = accuracy_score(y_test, y_pred)
print("Accuracy of Random Forest: ",acc_rf)
print()

#Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

#Classification Report
print(classification_report(y_test, y_pred))
```

```python
#Store Result in the list
storeResults('Random Forest', acc_rf)
```

```python
#hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt','log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 1000,10)]
# Minimum number of samples required to split a node
min_samples_split = [1,2, 5,10,14]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4,6,8]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
              'criterion':['entropy','gini']}
print(random_grid)
```

```python
#applying random_grid
rf1=RandomForestClassifier()
rf1_randomcv=RandomizedSearchCV(estimator=rf1,param_distributions=random_grid,n_iter=100,cv=3,verbose=2,
                               random_state=100,n_jobs=-1)
### fit the randomized model
rf1_randomcv.fit(X_train,y_train)

```

```python
rf1_randomcv.best_params_
```

```python
rf1_randomcv.best_estimator_
```

```python
best_random_grid=rf1_randomcv.best_estimator_
```

```python
y_pred = best_random_grid.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score {}".format(accuracy_score(y_test,y_pred)))
print("Classification report: {}".format(classification_report(y_test,y_pred)))
```

```python
#Gridsearch
from sklearn.model_selection import GridSearchCV

param_grid = {
    'criterion': [rf1_randomcv.best_params_['criterion']],
    'max_depth': [rf1_randomcv.best_params_['max_depth']],
    'max_features': [rf1_randomcv.best_params_['max_features']],
    'min_samples_leaf': [rf1_randomcv.best_params_['min_samples_leaf'], 
                         rf1_randomcv.best_params_['min_samples_leaf']+2, 
                         rf1_randomcv.best_params_['min_samples_leaf'] + 4],
    'min_samples_split': [rf1_randomcv.best_params_['min_samples_split'] - 2,
                          rf1_randomcv.best_params_['min_samples_split'] - 1,
                          rf1_randomcv.best_params_['min_samples_split'], 
                          rf1_randomcv.best_params_['min_samples_split'] +1,
                          rf1_randomcv.best_params_['min_samples_split'] + 2],
    'n_estimators': [rf1_randomcv.best_params_['n_estimators'] - 200, rf1_randomcv.best_params_['n_estimators'] - 100, 
                     rf1_randomcv.best_params_['n_estimators'], 
                     rf1_randomcv.best_params_['n_estimators'] + 100, rf1_randomcv.best_params_['n_estimators'] + 200]
}

print(param_grid)
```

```python
## Fit the grid_search to the data
rf1=RandomForestClassifier()
grid_search=GridSearchCV(estimator=rf1,param_grid=param_grid,cv=10,n_jobs=-1,verbose=2)
grid_search.fit(X_train,y_train)
```
